#### Median North American Video game sales

```python
n_american_median_sales = video_game_pd['NA_Sales'].median()
# n_american_median_sales = video_game_pd.median(10, 'NA_Sales')
n_american_median_sales
n_american_median_sales_10 = video_game_pd[video_game_pd['NA_Sales'] == 0.08][:10]
n_american_median_sales_10
```

#### North American top selling game of all time, # of standard deviations above/below mean of the sales

```python
n_american_sales_mean = video_game_pd['NA_Sales'].mean().round(decimals=3)
n_american_sales_mean
```

```python
n_american_sales_max = video_game_pd[video_game_pd['NA_Sales'] == video_game_pd['NA_Sales'].max()]
n_american_sales_max
```

```python
n_american_sales_std = video_game_pd['NA_Sales'].std()
n_american_sales_std
```

```python
n_american_zscore = (video_game_pd['NA_Sales'] - n_american_sales_mean) / video_game_pd['NA_Sales'].std()
n_american_zscore.head(1).round(decimals=2)
```
